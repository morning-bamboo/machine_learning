## K Means Clustering



```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
df = pd.read_excel("Lesson27-income.xlsx")
df
```

```python
plt.scatter(df.Age,df["Income($)"])
plt.xlabel("Age")
plt.ylabel("Income($)")
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(df[["Age","Income($)"]])

y_predicted
```

```python
df["cluster"] = y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1["Income($)"],color="red")
plt.scatter(df2.Age,df2["Income($)"],color="green")
plt.scatter(df3.Age,df3["Income($)"],color="blue")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="orange",marker="<",label="centroid")
plt.legend()
```

### Preprocessing using MinMaxScaler

```python
scaler = MinMaxScaler()

# scaler.fit(df[["Income($)"]])
df["Income($)"] = scaler.fit_transform(df[["Income($)"]])
df["Age"] = scaler.fit_transform(df[["Age"]])
df
```

```python
plt.scatter(df.Age,df["Income($)"])
plt.xlabel("Age")
plt.ylabel("Income($)")
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(df[["Age","Income($)"]])
y_predicted
```

```python
df["cluster"] = y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1["Income($)"],color="red")
plt.scatter(df2.Age,df2["Income($)"],color="green")
plt.scatter(df3.Age,df3["Income($)"],color="blue")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="orange",marker="<",label="centroid")
plt.xlabel("Age [Scaled]")
plt.ylabel("Income($) [Scaled]")

plt.legend()
```

### Elbow Plot (to help check optimal numbers of clusters)

```python
sse = [] # Sum of Squares Error 

k_range = range(1,10)
for k in k_range: 
    km = KMeans(n_clusters = k)
    km.fit(df[["Age","Income($)"]])
    sse.append(km.inertia_)

plt.plot(k_range, sse, linestyle="--",marker="o",color="red", label="Elbow Plot")
plt.xlabel("K")
plt.ylabel("Sum of Squares Error")
plt.legend()
```
