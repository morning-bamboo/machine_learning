### PCA (Principal Component Analysis) fitting involves transforming high-dimensional data into a lower-dimensional space by finding new, uncorrelated variables (Principal Components) that capture the maximum variance

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
import pandas as pd
from sklearn.decomposition import PCA
```

```python
n = 1000
pc1 = np.random.normal(0,1,n)
pc2 = np.random.normal(0,0.2,n)
noise = np.random.normal(0,0.05,n)

X = np.vstack((pc1+pc2,pc1-pc2,pc2+noise)).T
X.shape
```

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
vt = VarianceThreshold(0.05)
X_filtered = vt.fit_transform(X)
X_filtered.shape
```

```python
#draw image again from the above linear data: 

fig = plt.figure(figsize = (12,6))
#before PCA 
ax1 = fig.add_subplot(121,projection="3d")
ax1.scatter(X[:,0],X[:,1],X[:,2], color = "b")
plt.title("before PCA (3D)")


#after PCA
ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1], color = "r")
# ax2.set_xlim(-9,9)
# ax2.set_ylim(-2,2)
# ax2.set_xlim(-10, 10)
# ax2.set_ylim(-10, 10)
plt.title("after PCA (2D)")

plt.tight_layout()
plt.show()
```
