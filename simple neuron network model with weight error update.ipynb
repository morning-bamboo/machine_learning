# neuron network model





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_value = np.array([[0,0],[0,1],[1,0],[1,1]])

input_value
```

```python
output_value = np.array([0,1,1,1])
output_value
```

# output_value

```python
output_value = output_value.reshape(-1,1)

output_value
```

```python
weights = np.array([[0.1],[0.2]])

weights.shape
```

```python
bias = 0.3
```

```python
def sigmoid(xx):
    return 1/(1+np.exp(-xx))

def sigmoid_derivative(xx):
    return sigmoid(xx)*(1-sigmoid(xx))

xx = np.linspace(-10,10,1000) #used this specific example to visualize sigmoid & deri shape

y1 = sigmoid(xx)
y2 = sigmoid_derivative(xx)

plt.plot(xx,y1,label="sigmoid")
plt.plot(xx,y2,label="derivative")
plt.legend()
```

```python
# updateing weights and bias

j=0
k = []
l = []

learning_rate = 0.5

for epochs in range(2000):
    sum = np.dot(input_value,weights) + bias
    preds = sigmoid(sum)

    error = preds - output_value
    total_error = np.square(error).mean()  #mean-squared

    deriva_preds = sigmoid_derivative(preds) # backpropogation and chain rule
    deriva_product = deriva_preds*error 

    weights = weights - np.dot(input_value.T,deriva_product)*learning_rate # update weights by gradient descent algorithm


    j=j+1
    k.append(j)
    l.append(total_error)
    # print(j,total_error)
    for i in deriva_product:
        bias = bias - i*learning_rate
    
plt.plot(k,l)
weights,bias
```

```python
pred = np.array([1,1])

result = np.dot(pred,weights) + bias

res = sigmoid(result) >=1/2

res
```
